```python
import PyPDF2
import re

# Open PDF Resume
with open("VASUDEV.pdf", "rb") as file:
    pdf_reader = PyPDF2.PdfReader(file)

    text = ""

    # Read all pages
    for page in pdf_reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text

# Extract Email
email = re.findall(
    r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}",
    text
)

# Extract Phone Number
phone = re.findall(
    r"\b\d{10}\b",
    text
)

# Skills List
skills_database = [
    "Python",
    "Java",
    "C",
    "C++",
    "SQL",
    "Machine Learning",
    "Deep Learning",
    "HTML",
    "CSS",
    "JavaScript",
    "React",
    "Data Science"
]

# Find Skills
found_skills = []

for skill in skills_database:
    if skill.lower() in text.lower():
        found_skills.append(skill)

# Display Results
print("\n===== RESUME DETAILS =====")

if email:
    print("Email:", email[0])
else:
    print("Email: Not Found")

if phone:
    print("Phone:", phone[0])
else:
    print("Phone: Not Found")
    
print("Skills:", found_skills)
# Extract Projects Section
project_section = re.search(
    r"Projects(.*?)(Education|Skills|Experience|Certifications|Achievements|$)",
    text,
    re.IGNORECASE | re.DOTALL
)

project_titles = []

if project_section:
    projects_text = project_section.group(1)

    lines = projects_text.split("\n")

    for line in lines:
        line = line.strip()

        # Consider short lines as project titles
        if line and len(line.split()) <= 6:
            project_titles.append(line)

print("Project Titles:")
for i, title in enumerate(project_titles, 1):
    print(f"{i}. {title}")
```
